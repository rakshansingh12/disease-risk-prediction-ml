```python
from google.colab import drive
drive.mount('/content/drive')

import gzip

file_path = "/content/drive/MyDrive/disease-risk-prediction-gene-expression/data/raw/GSE5281_expression_matrix.txt.gz"

with gzip.open(file_path, "rt") as f:
    lines = f.readlines()
```

```python
unique_characteristics = set()

for line in lines:
    if "Sample_characteristics" in line:
        unique_characteristics.add(line.strip())

len(unique_characteristics)
```

```python
list(unique_characteristics)[:10]
```

```python
bio_source_lines = [
    line for line in lines
    if line.startswith("!Sample_characteristics_ch1")
    and "Bio-Source Name" in line
]

len(bio_source_lines)
```

```python
bio_source_line = bio_source_lines[0]

# Remove the leading metadata label
bio_source_line = bio_source_line.split("\t")[1:]

len(bio_source_line)
```

```python
labels = []

for entry in bio_source_line:
    e = entry.lower()

    if "control" in e:
        labels.append(0)
    elif "alzheimer" in e or " ad " in e or "affected" in e:
        labels.append(1)
    else:
        raise ValueError(f"Unmatched entry: {entry}")
```

```python
len(labels)
```

```python
labels.count(None)
```

```python
import numpy as np
y = np.array(labels)

np.unique(y, return_counts=True), y.shape
```

```python
np.save(
    "/content/drive/MyDrive/disease-risk-prediction-gene-expression/data/processed/y_labels.npy",
    y
)
```
